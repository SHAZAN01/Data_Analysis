```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [6,5,3,9,8,1]
```

```python
plt.xlabel('Days')
plt.ylabel('Stocks')
plt.plot(x,y)
plt.title('Stocks Vs Days')
plt.show()
```

```python
plt.xlabel('Days')
plt.ylabel('Stocks')
plt.plot(x,y,color='red',marker='*',mec='blue',alpha=0.2,linestyle='dashed',linewidth=1)
plt.title('Stocks Vs Days')
plt.figure(figsize=(15,25))
plt.show()
```

```python
import numpy as np
y = np.array([1,7,4,5,6,9,3])
z = np.array([11,17,14,15,16,19,13])
t = np.array([1,7,14,5,16,9,13])
plt.plot(y,marker='o',label='stars')
plt.plot(z,marker='*',label='celestials')
plt.plot(t,marker='p',label='planets')
plt.legend(shadow=True,loc='lower right')
plt.grid(axis='x')
plt.show()
```

Sub plots

Pie charts

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
e = [0.2,0.2,0.2,0.2]
plt.pie(y, labels = mylabels,startangle=90,explode=e, shadow=True)
plt.legend(title='fruits')
plt.title('colors')
plt.show()
```
